```python
def make_df_from_xml(path):
    xtree = et.parse(root_path + path)
    xroot = xtree.getroot()
    df_cols = ["conversation_id", "timestamp", "user", "text"]
    rows = []
    k = 0
    for node in xroot:
        if (node.attrib.get("conversation_id") != None):
            s_conversation_id = node.attrib.get("conversation_id")
            s_timestamp = node.find("ts").text
            s_user = node.find("user").text
            s_text = node.find("text").text
            rows.append({"conversation_id": s_conversation_id, "timestamp": s_timestamp, "user": s_user, "text": s_text})
        else:
            if (k==0):
                s_team_domain = node.find("team_domain")
                k = k + 1
            if (k==1):
                s_channel_name = node.find("channel_name")
                k = k + 1
            if (k==2):
                s_start_date = node.find("start_date")
                k = k + 1
            if (k==3):
                s_end_date = node.find("end_date")
                k = k + 1
    out_df = pd.DataFrame(rows, columns = df_cols)
    
    return out_df
    
```
